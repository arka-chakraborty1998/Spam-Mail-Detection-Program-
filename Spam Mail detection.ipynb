```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data=pd.read_csv("C:\\Users\\Acer\\Downloads\\mail_data.csv")
```

```python
raw_mail_data
```

```python
raw_mail_data.isnull().sum()
```

```python
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),' ')
mail_data
```

```python
mail_data.shape
```

```python

mail_data.loc[mail_data['Category']=='spam', 'Category']=0
mail_data.loc[mail_data['Category']=='ham', 'Category']=1
```

```python
mail_data
```

```python

X=mail_data['Message']
Y=mail_data['Category']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.3, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#calling the TfidfVectorizer function 
feature_extraction=TfidfVectorizer(min_df=1, stop_words="english", lowercase=True)
#now, we will fit test and train features in 
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)

#Converting Y_tarin and Y_test as integer datatype 
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
print(X_test_features)
```

```python
Y_train
```

```python
Y_test
```

```python
#calling the logistic regression model
model=LogisticRegression()
```

```python
#fitting the training features in the model
#X_train_features is refined form of X_train data
#Y_train is the integer form of Y-train data
model.fit(X_train_features, Y_train)
```

```python
#predicitng X_train data
prediction_on_training_data=model.predict(X_train_features)
#Matching the accuracy of Y_train with Predcited data of X_train
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
```

```python
accuracy_on_training_data
```

```python
#now let's do the prediction and accuracy check for test data also
model.fit(X_test_features, Y_test)
```

```python
#predicitng X_test data
prediction_on_testing_data=model.predict(X_test_features)
#Matching the accuracy of Y_train with Predcited data of X_train
accuracy_on_training_data=accuracy_score(Y_test,prediction_on_testing_data)
```
